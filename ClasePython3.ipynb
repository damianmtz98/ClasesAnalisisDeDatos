# **Cargar, filtrar y ordenar datos.**

```python
import pandas as pd
```

```python
df = pd.read_csv('usdmxn.csv')
```

```python
df.head(10)
```

```python
#Filtrar datos donde USD/MXN es menor que 18.8
filtro = df[df['USD/MXN'] < 18.8]
```

```python
print("Filtrado:")
print(filtro)
```

```python
#Ordenar por la columna 'USD/JPY' en orden descendente
ordenado = df.sort_values(by='USD/JPY', ascending= False)
```

```python
print(ordenado)
```

```python
# Calcular el promedio de 'USD/MXN' para cada valor Ãºnico en 'TasaDesempleoUSA'
promedio_usd_mxns = df.groupby('TasaDesempleoUSA')['USD/MXN'].mean()
```

```python
print("Promedio de USD/MXN por Tasa de Desempleo:")
print(promedio_usd_mxns)
```
